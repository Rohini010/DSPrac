```python
#Groupinh
group_mean=df.groupby('Product')['Value'].mean()
print(group_mean)

group_sum=df.groupby(['Product','Category'])['Value'].sum()
group_sum
```

```python
group_agg=df.groupby(['Category','Product'])['Sales'].agg(['mean','sum','count'])
group_agg
```

```python
#Merging & joining dataframe
df1=pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df1
```

```python
df2=pd.DataFrame({'Key':['A','B','D'], 'Value':[1,2,4]})
df2
```

```python

result_inner=pd.merge(df1,df2,on='Key',how='inner')
result_inner
```

```python
pd.merge(df1,df2,on='Key',how='outer')
```

```python
pd.merge(df1,df2,on='Key',how='left')
```

```python
pd.merge(df1,df2,on='Key',how='right')
```
